```python
import cv2
import numpy as np
from sklearn.cluster import MiniBatchKMeans
from sklearn.svm import LinearSVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, accuracy_score
from tqdm import tqdm

# Feature Extraction
def extract_features(image_path, method="SIFT", num_features=500):
    """
    Extract local features using SIFT or ORB.

    Args:
        image_path (str): Path to the image file.
        method (str): Feature extraction method ("SIFT" or "ORB").
        num_features (int): Number of features to extract.

    Returns:
        descriptors (ndarray): Local feature descriptors.
    """
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if image is None:
        print(f"Error loading image: {image_path}")
        return None

    if method == "SIFT":
        extractor = cv2.SIFT_create(nfeatures=num_features)
    elif method == "ORB":
        extractor = cv2.ORB_create(nfeatures=num_features)
    else:
        raise ValueError("Invalid method. Use 'SIFT' or 'ORB'.")

    keypoints, descriptors = extractor.detectAndCompute(image, None)
    if descriptors is None:
        print(f"No descriptors found for image: {image_path}")
    return descriptors

# Build BoW Model
def build_bow(descriptors_list, num_clusters=50):
    descriptors_list = [d for d in descriptors_list if d is not None]  # Filter None
    if len(descriptors_list) == 0:
        raise ValueError("No valid descriptors found.")

    all_descriptors = np.vstack(descriptors_list)
    kmeans = MiniBatchKMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(all_descriptors)
    return kmeans

# Extract BoW Features
def extract_bow_features(descriptors_list, kmeans, num_clusters=50):
    features = []
    for descriptors in descriptors_list:
        hist = np.zeros(num_clusters)
        if descriptors is not None:
            cluster_assignments = kmeans.predict(descriptors)
            for cluster_id in cluster_assignments:
                hist[cluster_id] += 1
        features.append(hist)
    return np.array(features)

# Train and Evaluate SVM
def train_and_evaluate_svm(train_features, train_labels, test_features, test_labels):
    scaler = StandardScaler()
    train_features = scaler.fit_transform(train_features)
    test_features = scaler.transform(test_features)

    svm = LinearSVC(random_state=42)
    svm.fit(train_features, train_labels)

    predictions = svm.predict(test_features)
    print("Classification Report:")
    print(classification_report(test_labels, predictions))
    print(f"Accuracy: {accuracy_score(test_labels, predictions):.2f}")
    return svm


# Execute the Pipeline
print("Extracting features...")
train_descriptors = [extract_features(file, method="SIFT") for file in tqdm(train_files)]
test_descriptors = [extract_features(file, method="SIFT") for file in tqdm(test_files)]

print("Building Bag-of-Words model...")
kmeans = build_bow(train_descriptors, num_clusters=50)

print("Extracting BoW features...")
train_features = extract_bow_features(train_descriptors, kmeans, num_clusters=50)
test_features = extract_bow_features(test_descriptors, kmeans, num_clusters=50)

print("Training and evaluating SVM...")
train_and_evaluate_svm(train_features, train_labels, test_features, test_labels)
```

```python
from torch.utils.data import DataLoader, Dataset
from PIL import Image
```

```python
class TinyImageNetDataset(Dataset):
    def __init__(self, image_paths, labels, transform=None):
        self.image_paths = image_paths
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        image = Image.open(self.image_paths[idx]).convert("RGB")
        label = self.labels[idx]
        if self.transform:
            image = self.transform(image)
        return image, label
```

```python
import torch
```

```python
# root_dir = "/kaggle/input/tinyimagenet/TinyImageNet100"  # Update this path as needed
# train_paths, test_paths, train_labels, test_labels = prepare_tinyimagenet(root_dir)

# Data Transforms
transform = transforms.Compose([
    transforms.Resize((64, 64)),
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

# Data Loaders
train_dataset = TinyImageNetDataset(train_files, train_labels, transform)
test_dataset = TinyImageNetDataset(test_files, test_labels, transform)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

print(f"Number of training samples: {len(train_dataset)}")
print(f"Number of testing samples: {len(test_dataset)}")

class FullyConnectedNN(torch.nn.Module):
    def __init__(self, input_size, num_classes):
        super(FullyConnectedNN, self).__init__()
        self.network = torch.nn.Sequential(
            torch.nn.Linear(input_size, 512),
            torch.nn.ReLU(),
            torch.nn.Linear(512, 256),
            torch.nn.ReLU(),
            torch.nn.Linear(256, num_classes),
        )

    def forward(self, x):
        return self.network(x)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Training and Evaluation for Fully Connected Neural Network
def train_and_evaluate_fc_with_metrics(train_loader, test_loader, input_size, num_classes, epochs=10):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = FullyConnectedNN(input_size, num_classes).to(device)
    criterion = torch.nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

    for epoch in range(epochs):
        model.train()
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            inputs = inputs.view(inputs.size(0), -1)  # Flatten inputs

            # Forward pass
            outputs = model(inputs)
            loss = criterion(outputs, labels)

            # Backward pass
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

        print(f"Epoch [{epoch + 1}/{epochs}], Loss: {loss.item():.4f}")

    # Evaluate
    model.eval()
    y_true, y_pred = [], []
    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            inputs = inputs.view(inputs.size(0), -1)  # Flatten inputs
            outputs = model(inputs)
            _, predicted = torch.max(outputs, 1)
            y_true.extend(labels.cpu().numpy())
            y_pred.extend(predicted.cpu().numpy())

    # Generate Confusion Matrix and Classification Report
    cm = confusion_matrix(y_true, y_pred)
    cr = classification_report(y_true, y_pred, target_names=[f"Class {i}" for i in range(num_classes)])
    print("Classification Report:\n", cr)

    # Plot Confusion Matrix
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=[f"Class {i}" for i in range(num_classes)],
                yticklabels=[f"Class {i}" for i in range(num_classes)])
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title("Confusion Matrix - Fully Connected Neural Network")
    plt.show()
```

```python
class SimpleCNN(torch.nn.Module):
    def __init__(self, num_classes):
        super(SimpleCNN, self).__init__()
        self.conv_layers = torch.nn.Sequential(
            torch.nn.Conv2d(3, 32, kernel_size=3, stride=1, padding=1),  # Output: (32, 64, 64)
            torch.nn.ReLU(),
            torch.nn.MaxPool2d(kernel_size=2, stride=2),                 # Output: (32, 32, 32)

            torch.nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1), # Output: (64, 32, 32)
            torch.nn.ReLU(),
            torch.nn.MaxPool2d(kernel_size=2, stride=2),                 # Output: (64, 16, 16)
        )
        self.fc_layers = torch.nn.Sequential(
            torch.nn.Linear(64 * 16 * 16, 256),
            torch.nn.ReLU(),
            torch.nn.Linear(256, num_classes),
        )

    def forward(self, x):
        x = self.conv_layers(x)
        x = x.view(x.size(0), -1)  # Flatten
        x = self.fc_layers(x)
        return x
```

```python
def train_and_evaluate_cnn_with_metrics(train_loader, test_loader, num_classes, epochs=10):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = SimpleCNN(num_classes).to(device)
    criterion = torch.nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

    for epoch in range(epochs):
        model.train()
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)

            # Forward pass
            outputs = model(inputs)
            loss = criterion(outputs, labels)

            # Backward pass
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

        print(f"Epoch [{epoch + 1}/{epochs}], Loss: {loss.item():.4f}")

    # Evaluate
    model.eval()
    y_true, y_pred = [], []
    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, predicted = torch.max(outputs, 1)
            y_true.extend(labels.cpu().numpy())
            y_pred.extend(predicted.cpu().numpy())

    # Generate Confusion Matrix and Classification Report
    cm = confusion_matrix(y_true, y_pred)
    cr = classification_report(y_true, y_pred, target_names=[f"Class {i}" for i in range(num_classes)])
    print("Classification Report:\n", cr)

    # Plot Confusion Matrix
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=[f"Class {i}" for i in range(num_classes)],
                yticklabels=[f"Class {i}" for i in range(num_classes)])
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title("Confusion Matrix - Convolutional Neural Network")
    plt.show()
```

```python
print("Phase 3: Fully Connected Neural Network with Metrics")
train_and_evaluate_fc_with_metrics(train_loader, test_loader, input_size=3*64*64, num_classes=15)
```

```python
print("\nPhase 4: Convolutional Neural Network with Metrics")
train_and_evaluate_cnn_with_metrics(train_loader, test_loader, num_classes=15)
```
